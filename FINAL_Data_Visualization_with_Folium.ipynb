# Step 2: Mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add the Marker cluster to the site map
site_map.add_child(marker_cluster)
for i in spacex_df.index:
    coor = [spacex_df['Lat'][i], spacex_df['Long'][i]]
    marker = folium.Marker(coor, icon=folium.Icon(color='white', icon_color=spacex_df['marker_color'][i]))
    marker_cluster.add_child(marker)
# for each row in spacex_df data frame
# create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successed or failed, e.g., icon=folium.Icon(color='white', icon_color=row['marker_color']
site_map
```

From the color-labeled markers in marker clusters, you should be able to easily identify which launch sites have relatively high success rates.


# Step 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now we zoom in to a launch site and explore its proximity to see if we can easily find any railway, highway, coastline, etc. We move the mouse to these points and mark down their coordinates (shown on the top-left) in order to calculate the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

Mark down a point on the closest railway using MousePosition and calculate the distance between the railway point to the launch site.


Coordinates of nearest railway point to VAFB SLC-4E launch site:
Lat: 34.63597
Long: -120.6239

```python
distance_railway = calculate_distance(lat1=34.63597, lon1=-120.6239, lat2=34.632834, lon2=-120.610746)
distance_railway
```

After obtained its coordinate, create a `folium.Marker` to show the distance


```python
# create and add a folium.Marker on your selected closest raiwaly point on the map
distmarker = folium.Marker([34.63597, -120.6239], icon=folium.Icon(icon='1.2533627411353925'), popup='1.2533627411353925')
site_map.add_child(distmarker)
site_map
# show the distance to the launch site using the icon property 
```

Draw a `PolyLine` between a launch site to the selected


```python
# Create a `folium.PolyLine` object using the railway point coordinate and launch site coordinate
poly = folium.PolyLine(locations=[[34.63597, -120.6239], [34.632834, -120.610746]])
site_map.add_child(poly)
site_map
```

Similarly, we can draw a line between a launch site to its closest city, coastline, highway, etc.


Drawing a line between KSC LC-39A launch site and the coastline nearest to it.
Nearest coastline coordinates:
Lat: 28.59897

```python
distance_coastline = calculate_distance(lat1=28.59897, lon1=-80.58579, lat2=28.573255, lon2=-80.646895)
distance_coastline
```

```python
distmarker2 = folium.Marker([28.59897, -80.58579], icon=folium.Icon(icon='6.618177826303218'), popup='6.618177826303218')
site_map.add_child(distmarker2)
poly2 = folium.PolyLine(locations=[[28.59897, -80.58579], [28.573255, -80.646895]])
site_map.add_child(poly2)
site_map
```

After we plot distance lines to the proximities, we can conclude:

*   All launch sites are always in close proximity to railways. This maybe because of the convenience of travelling.
*   The launch sites are not in close proximity to highways. This maybe for safety reasons.
*   Launch sites are not necessarily in close proximity to coastlines. CCAFS LC-40, CCAFS SLC-40 and VAFB SLC-4E are in close proximity to coastline but KSC LC-39A is not. The proximity to coastline may not have a significant impact on success/failures.
*   Yes, launch sites keep certain distance away from cities, over safety concerns.
